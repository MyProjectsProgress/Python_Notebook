## while with if-else

```python
#Guess the output
num = int(input())
while(num > 0):
    if num % 2 == 0:
        print ("The number you entered is now even")
        num = num - 1
    else:
        print ("The number you entered is now odd")
        num = num - 1
```

```python
#If the number of arguments is unknown, add a * list before the parameter name (*args)
def printNumbers(*numbers):
    for n in numbers:
        print(n)
printNumbers(1,2,5,11)
```

```python
#keyword arguments
def Teams(team1, team2, team3):
    print("The best team is " + team3)

Teams(team2 = "Al Ahly", team3 = "El Zamalek", team1 = "El Shar2ya lel Do5an")
```

```python
#If the number of keyword arguments is unknown, add ** dict before the parameter name (**kwargs)
def my_function(**me):
    print("My last name is " + me["lname"])

my_function(fname = "Abo", lname = "Youssef")
```

```python
#default parameters
def my_function(country = "Egypt"):
    print("I am from " + country)

my_function("Germany")
my_function()
```
